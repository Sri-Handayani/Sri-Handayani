### SOLUTION

```python
NEWTABLE = %sql select distinct C.ID ID\
,case when LASTNAME isnull then C.FIRSTNAME else C.FIRSTNAME||' '||C.LASTNAME end as NAME\
,C.COUNTRY as COUNTRY\
,case when C.PHONE isnull then '0' else '1' end as FLAG_NUMBER\
,coalesce(C.AGE,25) as AGE\
,coalesce(DAYS('2021-07-23')-DAYS(min(ORDERDATE) over (partition by C.ID)),0) as LOS\
, coalesce(sum(OI.TOTALPRODUCT) over (partition by C.ID),0) TOTALPRODUCT\
, coalesce(sum(OT.TOTALAMOUNT) over (partition by C.ID),0) TOTALAMOUNT\
, coalesce(min(OT.PAYMENTTYPE) over (partition by C.ID),1 ) PAYMENTTYPE\
from CUSTOMER C left outer join \
(select ID,CUSTOMERID,ORDERDATE,TOTALAMOUNT,\
case when PAYMENTTYPE = 'Cash' or PAYMENTTYPE= null then '1' else '0' end as PAYMENTTYPE \
from ORDER_TRANSACTION) OT\
on C.ID=OT.CUSTOMERID\
left outer join \
(select ORDERID,sum(QUANTITY) as TOTALPRODUCT from ORDERITEM group by ORDERID) OI\
on OT.ID=OI.ORDERID\
order by C.ID;
```
